```python
# standard
import numpy as np
import pandas as pd
from numpy.random import randn

#stats
from scipy import stats

#plotting
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
# allows us to see plots
```

```python
#plotting a univariate distribution along the X axis
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
tips.tail()
```

```python
sns.kdeplot(data=tips, x='total_bill')
```

```python
sns.kdeplot(data=tips, x='total_bill',color='green',shade=True)
```

```python
#flipping the plot assigning data variable to y axis
sns.kdeplot(data=tips, y='total_bill')
```

```python
# using less smoothing
sns.kdeplot(data=tips, x='total_bill', bw_adjust=.2)
```

```python
# using more smoothing but dont smooth past the extreme data points
ax=sns.kdeplot(data=tips, x='total_bill', bw_adjust=5, cut=0)
```

```python
# plotting a conditional distribution with hue mapping of a second variable
sns.kdeplot(data=tips, x='total_bill', hue='time')
```

```python
# stack the conditional distributions
sns.kdeplot(data=tips, x='total_bill', hue='time', multiple='stack')
```

```python
# normalizing the stacked distribution at each value in the grid
sns.kdeplot(data=tips, x='total_bill', hue='time', multiple='fill')
```

```python
#Estimating the cumulative distribution function(s),normalizing each subset
sns.kdeplot(data=tips, x='total_bill', hue='time',
           cumulative=True, common_norm=False, common_grid=True)
```

```python
# Estimating distribution from aggregated data,using weights
tips_agg = (tips.groupby('size').agg(total_bill=('total_bill', 'mean'), n=('total_bill', 'count')))
sns.kdeplot(data=tips_agg, x='total_bill', weights='n')
```

```python
# Using numeric hue mapping
sns.kdeplot(data=tips, x='total_bill', hue='size')
```

```python
# modifying appearance of the plot
sns.kdeplot(data=tips, x='total_bill', hue='size', fill=True, common_norm=False,
           palette='crest',alpha=.5, linewidth=0)
```

looking into another dataset


```python
geyser = sns.load_dataset('geyser')
```

```python
geyser.head()
```

```python
#plotting a bivariate distribution
sns.kdeplot(data=geyser, x='waiting', y='duration')
```

```python
# mapping a third variable with a hue semantic showing conditional distribution
sns.kdeplot(data=geyser, x='waiting', y='duration', hue='kind')
```

```python
#showing filled contours
sns.kdeplot(data=geyser, x='waiting', y='duration', hue='kind', fill=True)
```

```python
# Showing fewer contour levels,covering lesss of the distribution
sns.kdeplot(data=geyser, x='waiting', y='duration', hue='kind', levels=5, thresh=.2)
```

```python
# filling the axes extent with a smooth distribution ,using a different colormap
sns.kdeplot(data=geyser, x='waiting', y='duration', hue='kind', fill=True, thresh=0,
           levels=100)
```

```python
sns.jointplot(data=geyser,x='waiting',y='duration',kind='kde')
```
